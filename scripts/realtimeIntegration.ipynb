```python
import cv2
from ultralytics import YOLO

# Load the trained YOLO model
model_path = "/home/shashank/Desktop/Mini_project /runs/detect/pothole_detection9/weights/best.pt"  # Replace with your model path
model = YOLO(model_path)

# Set the webcam index for DroidCam
webcam_index = 10 # Adjust as needed
cap = cv2.VideoCapture(webcam_index)

if not cap.isOpened():
    print(f"Error: Unable to open webcam at index {webcam_index}.")
    exit()

print("Starting real-time pothole detection. Press 'Ctrl+C' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Unable to read from the webcam. Exiting...")
        break

    # Perform inference
    results = model.predict(source=frame, save=False, imgsz=640)

    # Save annotated frame to disk
    annotated_frame = results[0].plot()
    cv2.imwrite("output.jpg", annotated_frame)

    # Optionally log progress
    print("Frame saved as output.jpg")

cap.release()
```

```python
import cv2
from ultralytics import YOLO

# Load the trained YOLO model
model_path = "/home/shashank/Desktop/Mini_project /runs/detect/pothole_detection9/weights/best.pt"  # Replace with your model path
model = YOLO(model_path)

# Set the webcam index for DroidCam
webcam_index = 10  
cap = cv2.VideoCapture(webcam_index)

if not cap.isOpened():
    print(f"Error: Unable to open webcam at index {webcam_index}.")
    exit()
print("Starting real-time pothole detection. Press 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Unable to read from the webcam. Exiting...")
        break

    # Perform inference
    results = model.predict(source=frame, save=False, imgsz=640)

    # Annotate the frame with detections
    annotated_frame = results[0].plot()

    # Display the annotated frame in a window
    cv2.imshow("Real-Time Pothole Detection", annotated_frame)

    # Exit the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close the OpenCV window
cap.release()
cv2.destroyAllWindows()
```
